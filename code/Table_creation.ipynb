```python
import pandas as pd 
import numpy as np

from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier

from numpy import mean
from numpy import std

import sklearn
from sklearn.metrics import accuracy_score,confusion_matrix, roc_curve
```

```python
ddi = pd.read_csv("ddi_drugbank_final.txt", sep="\t", header=None)
chemical = pd.read_csv("DRSim_chem_drugbank_final.txt", sep="\t", header=None)
atc = pd.read_csv("DRSim_atc_db_final.txt", sep="\t", header=None)
anatomical = pd.read_csv("anatomical_similarity.txt", sep="\t", header=None)
indication = pd.read_csv("DRSim_indication_sider_ctd_final.txt",sep="\t", header=None)
se = pd.read_csv("DRSim_se_sider_final.txt",sep="\t", header=None)
target=pd.read_csv("DRSim_target_drugbank_final.txt",sep="\t", header=None)
enzyme =pd.read_csv("DRSim_target_drugbank_final.txt",sep="\t", header=None)
```

```python
##atc code
atc_dict = {}
atc_drug = []

for line in range(len(atc)):
    db1 = atc[0][line]
    db2 = atc[1][line]
    sim = atc[2][line]
    atc_drug.append(db1)
    atc_drug.append(db2)
      
    if db1 in atc_dict.keys():
        atc_dict[db1][db2]=sim
    else :
        atc_dict[db1] = {}
        atc_dict[db1][db2]=sim   
```

```python
#chem
chemical_dict ={}
chemical_drug = []

for line in range(len(chemical)):
    db1 = chemical[0][line]
    db2 = chemical[1][line]
    sim = chemical[2][line]
    chemical_drug.append(db1)
    chemical_drug.append(db2)
      
    if db1 in chemical_dict.keys():
        chemical_dict[db1][db2]=sim
    else :
        chemical_dict[db1] = {}
        chemical_dict[db1][db2]=sim  
```

```python
#indication
indication_dict ={}
indication_drug = []

for line in range(len(indication)):
    db1 = indication[0][line]
    db2 = indication[1][line]
    sim = indication[2][line]
    indication_drug.append(db1)
    indication_drug.append(db2)
      
    if db1 in indication_dict.keys():
        indication_dict[db1][db2]=sim
    else :
        indication_dict[db1] = {}
        indication_dict[db1][db2]=sim  
```

```python
#se
se_dict ={}
se_drug =[]

for line in range(len(se)):
    db1 = se[0][line]
    db2 = se[1][line]
    sim = se[2][line]
    se_drug.append(db1)
    se_drug.append(db2)
      
    if db1 in se_dict.keys():
        se_dict[db1][db2]=sim
    else :
        se_dict[db1] = {}
        se_dict[db1][db2]=sim 
```

```python
#target
target_dict ={}
target_drug = []

for line in range(len(target)):
    db1 = target[0][line]
    db2 = target[1][line]
    sim = target[2][line]
    target_drug.append(db1)
    target_drug.append(db2) 
    
    if db1 in target_dict.keys():
        target_dict[db1][db2]=sim
    else :
        target_dict[db1] = {}
        target_dict[db1][db2]=sim  
```

```python
#target
anatomical_dict ={}
anatomical_drug = []

for line in range(len(anatomical)):
    db1 = anatomical[0][line]
    db2 = anatomical[1][line]
    sim = anatomical[2][line]
    anatomical_drug.append(db1)
    anatomical_drug.append(db2) 
    
    if db1 in anatomical_dict.keys():
        anatomical_dict[db1][db2]=sim
    else :
        anatomical_dict[db1] = {}
        anatomical_dict[db1][db2]=sim  
```

```python
#ddi
ddi_dict ={}
ddi_drug = []

for line in range(len(ddi)):
    db1 = ddi[0][line]
    db2 = ddi[1][line]
    sim = 1
    ddi_drug.append(db1)
    ddi_drug.append(db2)
    if db1 in ddi_dict.keys():
        ddi_dict[db1][db2]=sim
    else :
        ddi_dict[db1] = {}
        ddi_dict[db1][db2]=sim  
```

```python
#enzyme 
enzyme_dict = {}
enzyme_drug = [] 

for line in range(len(enzyme)):
    db1 = enzyme[0][line]
    db2 = enzyme[1][line]
    sim = enzyme[2][line]
    enzyme_drug.append(db1)
    enzyme_drug.append(db2)
      
    if db1 in enzyme_dict.keys():
        enzyme_dict[db1][db2]=sim
    else :
        enzyme_dict[db1] = {}
        enzyme_dict[db1][db2]=sim   
```

```python
#### common drug from the properties --> and make pairs 
common_drug = list(set(chemical_drug) &set(enzyme_drug) & set(target_drug)  & set(atc_drug)&set(se_drug) & set(indication_drug)  & set(ddi_drug) & set(anatomical_drug))
drug_list = pd.DataFrame(common_drug)
whole_list = []
for index1,row1 in drug_list.iterrows():
    for index2, row2 in drug_list.iterrows():
        mid = []
        if row1[0] < row2[0]:
            mid.append(row1[0])
            mid.append(row2[0])
            whole_list.append(mid)
            type = ddi_dict[row1[0]].get(row2[0],0 )
            mid.append(type)
whole_list = pd.DataFrame(whole_list)
```

```python
whole_list.columns =['Drug1','Drug2','Class']
```
