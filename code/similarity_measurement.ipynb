```python
#import libraries
import pandas as pd
import numpy as np

## change the file name below (according to input)
to_calculate = pd.read_csv("atc_drugbank_final.txt", sep="\t", header = None)
```

```python
#Store data to dictionary
#format is like drug name (key), properties related to drugs (value)
data_dict = {}

for line in range(len(to_calculate)):
    db1 = to_calculate[0][line]
    sim = to_calculate[1][line]
      
        
    if db1 in data_dict.keys():
        data_dict[db1].append(sim)
    else :
        data_dict.setdefault(db1, [])
        data_dict[db1].append(sim)
        
```

```python
data_dict
```

```python
# define 
def intersection (list1, list2):
    return list(set(list1) & set(list2))
```

```python
def union (list1, list2):
    return list(set(list1) | set(list2))
```

```python
similarity =[]

for keys1 in data_dict:
    lst1 = data_dict[keys1]
    for keys2 in data_dict:
        if keys1 !=  keys2 : 
            lst2= data_dict[keys2] 
            on_fraction = intersection(lst1, lst2)
            under_fraction = union(lst1, lst2)
            sim_val = float(len(on_fraction)/len(under_fraction))
            mid1 = [keys1, keys2, sim_val]
            mid2 = [keys2, keys1, sim_val]
            similarity.append(mid1)
            similarity.append(mid2)
            
```

```python
similarity = pd.DataFrame(similarity)
similarity = similarity.drop_duplicates()
```
